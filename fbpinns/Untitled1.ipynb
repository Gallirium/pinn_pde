```python
subdomain_xs = [jnp.linspace(0, 3, 12), jnp.linspace(0, 1, 8)]
subdomain_ws = get_subdomain_ws(subdomain_xs, 2.7)

c = Constants(
    run = "TEST_KDV_FBPINN_0",
    domain=RectangularDomainND,
    domain_init_kwargs=dict(
        xmin=np.array([0., 0.]),
        xmax=np.array([3., 1.]),
    ),
    problem=KdV,
    problem_init_kwargs=dict(c=4, a=0),
    decomposition = RectangularDecompositionND,
    decomposition_init_kwargs=dict(
        subdomain_xs = subdomain_xs,
        subdomain_ws = subdomain_ws,
        unnorm=(0., 1.),
    ),
    network=FCN,
    network_init_kwargs=dict(
        layer_sizes = [2,32,1],
    ),
    # scheduler = LineSchedulerRectangularND,
    # scheduler_kwargs = dict(
    #     point=[0.], iaxis=0,
    # ),
    ns = ((100,100),),
    n_test = (200,200),
    n_steps = 100000,
    optimiser_kwargs=dict(learning_rate=1e-4),
    show_figures=True,
    clear_output=True);
```
